```python
#W6A9
d1 = {}
d2 = {}
n = int(input("Nhập số cặp key, value của dictionary 1: "))
for _ in range(n):
    
    k, v = map(str, input("Nhập key, value cách nhau bởi dấu hai chấm: ").split(':'))
    d1[k] = v
n = int(input("Nhập số cặp key, value của dictionary 2: "))
for _ in range(n):
    k, v = map(str, input("Nhập key, value cách nhau bởi dấu hai chấm: ").split(':'))
    d2[k] = v
for k, v in d2.items():
    if k in d1:
        d1[k] = [d1[k], v]
    else:
        d1[k] = [v]
a = []
for k, v in d1.items():
    a.append([k, v])
a.sort()
print("{", end = '')
for i in range(len(a)):
    x = a[i]
    if i < len(a) - 1:
        print(f'\'{x[0]}\': \'{x[1]}\', ', end = '')
    else:
        print(f'\'{x[0]}\': \'{x[1]}\'', end = '')
print('}')
```

```python
ans = []
a = list(map(int, input().split()))
k = int(input())
n = len(a)
for i in range(n):
    for j in range(i + 1, n):
        if a[i] + a[j] == k:
            ans.append((i + 1, j + 1))
ans.sort()
print(ans)
```

```python
best = 0
num = 0
d = dict()
for x in map(int, input().split()):
    d[x] = d.get(x, 0) + 1
for k, v in d.items():
    if v > best:
        best = v
        num = k
    elif v == best:
        num = min(num, k)
print(num)
```
