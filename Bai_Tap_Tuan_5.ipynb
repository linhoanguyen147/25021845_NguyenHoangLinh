```python
#W5A1
def Max_Of_Two(a, b) -> int:
    return a if a > b else b
print(Max_Of_Two(10, 20))
```

```python
#W5A2
def swap(a, b) -> tuple[int, int]:
    c = a
    a = b
    b = c
    return (a, b)
print(*swap(2, 5))
```

```python
#W5A3
def isPrime(x) -> bool:
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True
print(isPrime(17))
```

```python
#W5A4
def isPerf(x) -> bool:
    res = 1
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            res += i
            if i * i != x:
                res += x // i
    return res == x
print(isPerf(6))
```

```python
#w5A5
def findK(a: list[int], k: int) -> int:
    for i in range(len(a)):
        if a[i] == k:
            return i   
    return -1
print(findK([1, 5, 3, 2, 7], 10))
```

```python
#W5A6
def fac(n: int) -> int:
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res
print(fac(10))
```

```python
#W5A7
def calc(num1: int, num2: int, op: str):
    if op == '+':
        return num1 + num2
    if op == '-':
        return num1 - num2
    if op == '*':
        return num1 * num2
    if op == '/':
        return num1 / num2
print(calc(5, 10, '-'))
```

```python
#W5A8
def Hamming(a, b):
    c = a ^ b
    cnt = 0
    while c:
        if c & 1:
            cnt += 1
        c >>= 1
    return cnt
print(Hamming(1, 4))
```

```python
#W5A9
def digitSum(n):
    res = 0
    while n:
        res += n % 10
        n //= 10
    return res
print(digitSum(12351384))
```

```python
#W5A10
def check(a, b):
    d = dict()
    cnt = 0
    h1 = 0
    for i in range(len(a)):
        if a[i] in d:
            pass
        else:
            cnt += 1
            d[a[i]] = cnt
        h1 = h1 * 10 + d[a[i]]
    h2 = 0
    cnt = 0
    d = dict()
    for i in range(len(b)):
        if b[i] in d:
            pass
        else:
            cnt += 1
            d[b[i]] = cnt
        h2 = h2 * 10 + d[b[i]]
    return h1 == h2
print(check("abca", "zbxz"))
```
