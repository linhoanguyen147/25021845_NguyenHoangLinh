```python
file1 = "name1.txt"
file2 = "name2.txt"
with open(file1, 'w') as f: f.write("Tom Marvolo Riddle")
with open(file2, 'w') as f: f.write("I Am Lord Voldemort")

def areAnagrams(path1: str, path2: str) -> bool:
    def clean_str(s):
        return "".join(sorted(s.lower().replace(" ", "")))

    try:
        with open(path1, 'r') as f1, open(path2, 'r') as f2:
            s1 = f1.read().strip()
            s2 = f2.read().strip()
            return clean_str(s1) == clean_str(s2)
    except FileNotFoundError:
        return False

print(areAnagrams(file1, file2))
```

```python
def fibonacci_words(n):
    if n == 0: return ["a"]
    if n == 1: return ["a", "b"]
    
    res = ["a", "b"]
    for i in range(2, n + 1):
        next_val = res[i-1] + res[i-2]
        res.append(next_val)
        
    with open('output.txt', 'w') as f:
        for idx, val in enumerate(res):
            f.write(f"f({idx}) = {val}\n")
    
    for idx, val in enumerate(res):
        print(f"f({idx}) = {val}")

fibonacci_words(6)
```

```python
filename = "houses.txt"
with open(filename, 'w') as f: f.write("1 2 3 1")

def getMoney(path: str) -> list[int]:
    try:
        with open(path, 'r') as f:
            content = f.read().split()
            return [int(x) for x in content]
    except:
        return []

def rob(nums: list[int]) -> int:
    if not nums: return 0
    if len(nums) == 1: return nums[0]
    
    prev1 = 0
    prev2 = 0
    
    for num in nums:
        tmp = prev1
        prev1 = max(prev2 + num, prev1)
        prev2 = tmp
    return prev1

houses = getMoney(filename)
print(f"Danh sách tiền các nhà: {houses}")
print(f"Số tiền tối đa trộm được: {rob(houses)}")
```

```python
filename = "max_prod.txt"
with open(filename, 'w') as f: 
    f.write("5\n-10 -10 5 2 3") 

def maximumProduct(path: str) -> int:
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            nums = list(map(int, lines[1].split()))
            
            nums.sort()
            
            return max(nums[-1] * nums[-2] * nums[-3], nums[0] * nums[1] * nums[-1])
    except:
        return 0

print(maximumProduct(filename))
```

```python
filename = "zeroes.txt"
with open(filename, 'w') as f: 
    f.write("5\n0 1 0 3 12")

def moveZeroes(fileName: str) -> list:
    try:
        with open(fileName, 'r') as f:
            lines = f.readlines()
            nums = list(map(int, lines[1].split()))
            
            non_zero = [x for x in nums if x != 0]
            count_zero = len(nums) - len(non_zero)
            
            return non_zero + [0] * count_zero
    except:
        return []

print(moveZeroes(filename))
```

```python
import csv

filename = "racing.csv"
content = """Lap,Time
1,143.23
2,134.12
3,150.05"""
with open(filename, 'w') as f: f.write(content)

def averageTime(path: str):
    try:
        with open(path, 'r') as f:
            reader = csv.reader(f)
            next(reader) 
            
            total_time = 0
            count = 0
            
            for row in reader:
                if row:
                    total_time += float(row[1])
                    count += 1
            
            if count > 0:
                return round(total_time / count, 2)
            return 0.0
    except Exception as e:
        print(e)
        return 0.0

print(averageTime(filename))
```

```python
filename = "movies.txt"
content = """id, movie, description, rating
1, Thor, great 3D, 8.0
2, The Rings, Fantasy, 7.5
3, Irish, boring, 6.2
4, Ice song, Fantasy, 8.6
5, Home Alone, Interesting, 6.5"""
with open(filename, 'w') as f: f.write(content)

def findMovies(path: str) -> list[str]:
    result = []
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            for line in lines[1:]:
                parts = line.strip().split(',')
                if len(parts) >= 4:
                    movie_name = parts[1].strip()
                    desc = parts[2].strip().lower()
                    rating = float(parts[3].strip())
                    
                    if rating > 8.0 and "boring" not in desc:
                        result.append(movie_name)
        
        if not result:
            return ["All are boring!"]
        return result
    except:
        return []

movies = findMovies(filename)
for m in movies:
    print(m)
```

```python
import math

filename = "shapes.txt"
content = """TYPE: SQUARE
SIDE: 5.3

TYPE: RECTANGLE
WIDTH: 5.0 HEIGHT: 16.0

TYPE: CIRCLE
RADIUS: 2.15"""
with open(filename, 'w') as f: f.write(content)

def calculate_perimeters(path):
    try:
        with open(path, 'r') as f:
            lines = [line.strip() for line in f if line.strip()] 
        
        i = 0
        while i < len(lines):
            line = lines[i]
            if line.startswith("TYPE:"):
                shape_type = line.split(":")[1].strip()
                i += 1
                params_line = lines[i]
                
                perimeter = 0
                if shape_type == "SQUARE":
                    side = float(params_line.split(":")[1])
                    perimeter = 4 * side
                    
                elif shape_type == "RECTANGLE":
                    parts = params_line.split()
                    width = float(parts[1])
                    height = float(parts[3])
                    perimeter = 2 * (width + height)
                    
                elif shape_type == "CIRCLE":
                    radius = float(params_line.split(":")[1])
                    perimeter = 2 * math.pi * radius
                
                print(f"{perimeter:.2f}")
            i += 1

    except Exception as e:
        print(f"Error: {e}")

path_input = input("Nhập đường dẫn file (VD: shapes.txt): ")
calculate_perimeters(path_input)
```

```python
filename = "array_prod.txt"
with open(filename, 'w') as f: f.write("5\n4 6 2 5 2")

def productExceptSelf(path: str) -> list[int]:
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            nums = list(map(int, lines[1].split()))
            n = len(nums)
            
            result = [1] * n
            
            left = 1
            for i in range(n):
                result[i] = left
                left *= nums[i]
            
            right = 1
            for i in range(n - 1, -1, -1):
                result[i] *= right
                right *= nums[i]
                
            return result
    except:
        return []

print(*productExceptSelf(filename))
```

```python
import math

filename = "rocket.txt"
content = """SRC X: 0 Y: 0
DES X: 10000 Y: 10000
VELOCITY: 6 KM/S"""
with open(filename, 'w') as f: f.write(content)

def estimatedTime(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            
            src_parts = lines[0].split()
            src_x = float(src_parts[2])
            src_y = float(src_parts[4])
            
            des_parts = lines[1].split()
            des_x = float(des_parts[2])
            des_y = float(des_parts[4])
            
            vel_parts = lines[2].split()
            velocity = float(vel_parts[1])
            
            distance = math.sqrt((des_x - src_x)**2 + (des_y - src_y)**2)
            
            time = distance / velocity
            return round(time, 2)
    except Exception as e:
        print(e)
        return 0.0

print(estimatedTime(filename))
```

```python
import struct

filename = "data.bin"
data = [1.234567, 5.555555, 9.999999]
with open(filename, 'wb') as f:
    for num in data:
        f.write(struct.pack('d', num))

path = input("Nhập đường dẫn file binary (VD: data.bin): ")

try:
    with open(path, 'rb') as f:
        content = f.read()
        num_doubles = len(content) // 8
        
        numbers = struct.unpack('d' * num_doubles, content)
        
        for num in numbers:
            print(f"{num:.5f}")
except FileNotFoundError:
    print("Không tìm thấy file")
except Exception as e:
    print(f"Lỗi: {e}")
```

```python
import math

filename = "image.txt"
content = """4
130 160 150 170
160 140 180 160
170 190 120 150
210 180 160 120"""
with open(filename, 'w') as f: f.write(content)

def imageSmoother(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            M = int(lines[0].strip())
            img = []
            for i in range(M):
                row = list(map(int, lines[i+1].split()))
                img.append(row)
        
        res = [[0] * M for _ in range(M)]
        
        for r in range(M):
            for c in range(M):
                total = 0
                count = 0
                for i in (r-1, r, r+1):
                    for j in (c-1, c, c+1):
                        if 0 <= i < M and 0 <= j < M:
                            total += img[i][j]
                            count += 1
                
                res[r][c] = math.floor(total / count)
        
        for row in res:
            print(*(row))
            
    except Exception as e:
        print(f"Error: {e}")

imageSmoother(filename)
```
