```python
n = int(input())
a = []
for i in range(n):
    a.append(list(map(int, input().split())))
for i in range(n):
    print(a[i][i], end = ' ')
print()
for i in range(n):
    print(a[i][n - i - 1], end = ' ')
```

```python
n, m, k = map(int, input().split())
a = []
for i in range(n):
    a.append(list(map(int, input().split())))
s = 0
for i in range(n):
    s += a[i][k - 1]
print(s)
s = 0
for j in range(m):
    s += a[k - 1][j]
print(s)
```

```python
a = [int(x) for x in input().split()]
mx = max(a)
for i, val in enumerate(a):
    if val == mx:
        print(1 + i)
        break
```

```python
a = [int(x) for x in input().split()]
k = int(input())
for i in range(len(a)):
    if a[i] == k:
        print(i + 1)
        break
else:
    print(-1)
```

```python
a = [int(x) for x in input().split()]
cnt = [0] * 10
for i in a:
    cnt[i] = 1
for i in range(10):
    if not cnt[i]:
        print(i)
```

```python
a = [int(x) for x in input().split()]
s = 0
for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        s += a[i]
print(s)
```

```python
a = [int(x) for x in input().split()]
x = int(input())
b = [i + 1 for i, z in enumerate(a) if z < x]
if len(b):
    print(*b)
else:
    print(-1)
```

```python
a = [int(x) for x in input().split()]
tang = 1
for i in range(1, len(a)):
    if a[i] <= a[i - 1]:
        tang = 0
        break
if tang:
    print("Tang")
else:
    giam = 1
    for i in range(1, len(a)):
        if a[i] >= a[i - 1]:
            giam = 0
    if giam:
        print("Giam")
    else:
        print("Vo danh")
```
