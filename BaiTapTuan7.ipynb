```python
def binary_search(arr, left, right, target):
    l = left
    r = right
    while l <= r:
        mid = l + r >> 1
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            l = mid + 1
        else:
            r = mid - 1
    return -1
arr = [int(x) for x in input().split()]
x = int(input())
print(binary_search(arr, 0, len(arr) - 1, x))
```

```python
def count_occurrences(arr, x) -> int:
    arr.sort()
    cnt = 0
    for i in range(len(arr)):
        if arr[i] == x:
            for j in range(i, len(arr)):
                if arr[j] == x:
                    cnt += 1
            break
    return cnt
arr = [int(x) for x in input().split()]
x = int(input())
print(count_occurrences(arr, x))
```
